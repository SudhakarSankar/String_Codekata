```python
# 5
'''you are given a string made up of parenthesis only.Your task is to check whether parenthesis are balanced or not.
If they are balanced print 1 else print 0'''
# Input Description:
# You are given a string ‘s’
# Output Description:
# Print 1 for balanced and 0 for imbalanced

# Sample Input :
# {({})}
# Sample Output :
# 1

def is_balanced(s):
    stack = []
    matching = {')': '(', '}': '{', ']': '['}

    for char in s:
        if char in "({[":
            stack.append(char)
        elif char in ")}]":
            if not stack or stack.pop() != matching[char]:
                return 0  # Imbalanced
    
    return 1 if not stack else 0  # Stack should be empty for balance

# Input from user
s = input().strip()
print(is_balanced(s))
```

```python
# 13
# You are given a number with duplicate digits your task is to remove the immediate duplicate digits and print the result
# Input Description:
# You are given a long string of digits
# Output Description:
# Print the desired result print -1 if result length is 0

# Sample Input :
# 1331
# Sample Output :
# 11

def remove_immediate_duplicates(digits):
    result = []  # Using a list as a stack
    
    for digit in digits:
        if result and result[-1] == digit:
            result.pop()  # Remove the last element if it's a duplicate
        else:
            result.append(digit)  # Add the digit if it's not an adjacent duplicate

    return ''.join(result) if result else '-1'

# Taking input
digits = input().strip()

# Output result
print(remove_immediate_duplicates(digits))
```

```python
# 30
'''Given an array of pairs of strings, find if there are mirror pairs.
(s1, s2) & (s3, s4) are mirror pairs, if s1 = s4 and s2 = s3. The first string in each pair is distinct.'''
# Input Description:
# The first line contains the number string pairs N. Then N string pairs follow.
# Output Description:
# Print YES, if a mirror pair exists, print NO otherwise.

# Sample Input :
# 3
# raja kili
# pan quil
# kili raja
# Sample Output :
# YES

def has_mirror_pair(n, pairs):
    pair_dict = {}
    for s1, s2 in pairs:
        if (s2, s1) in pair_dict:
            print("YES")
            return
        pair_dict[(s1, s2)] = True
    print("NO")

# Sample Input
n = int(input().strip())
pairs = [tuple(input().strip().split()) for _ in range(n)]
has_mirror_pair(n, pairs)
```

```python
# 31
# Given a number ‘n’, your task is to  generate all n-bit grey code sequences, “a grey code sequence is a sequence 
# such that successive patterns in it differ by one bit”
# Input Description:
# You are given an number ‘n’.
# Output Description:
# Print the grey code sequence

# Sample Input :
# 2
# Sample Output :
# 00 01 11 10

def generate_gray_code(n):
    result = []
    for i in range(2 ** n):
        gray_code = i ^ (i >> 1)  # Generate gray code
        result.append(f"{gray_code:0{n}b}")  # Format as binary with leading zeros
    print(" ".join(result))

# Sample Input
n = int(input().strip())
generate_gray_code(n)
```

```python
# 33
'''You are given two strings . Your task is to tell whether the pair of strings is panagram.
A pair of strings are said to be panagram if they both are palindrome and are anagram of each other.'''
# Input Description:
# You will be given two strings ‘s1’ and ‘s2’
# Output Description:
# Print 1 if they are panagram and 0 if they are not

# Sample Input :
# nitin intni
# Sample Output :
# 1

def is_palindrome(s):
    return s == s[::-1]

def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)

def is_panagram(s1, s2):
    if is_palindrome(s1) and is_palindrome(s2) and is_anagram(s1, s2):
        print(1)
    else:
        print(0)

# Sample Input
s1, s2 = input().strip().split()
is_panagram(s1, s2)
```

```python
# 34
'''In a school there are voting to choose the monitor of class.Your task is to tell which candidate is winner 
and if there is a tie print the name of candidate whose come first in lexicographical order.'''
# Input Description:
# You are given with the space separated names.
# Output Description:
# Print the winner’s name and the votes he earned.

# Sample Input :
# john johnny jackie johnny john jackie jamie jamie john johnny jamie johnny john
# Sample Output :
# john 4
from collections import Counter

def find_winner(votes):
    vote_count = Counter(votes)  # Count occurrences of each name
    max_votes = max(vote_count.values())  # Find the maximum votes
    
    # Get the candidate(s) with the maximum votes
    winner = min([name for name, count in vote_count.items() if count == max_votes])
    print(winner, max_votes)

# Sample Input
votes = input().strip().split()
find_winner(votes)

```

```python
# 36
'''Guvi developed a new system to make sure no two usernames are same. So, they hired you as a developer to develop this system. 
They have set some rules to do the same.If you see the same username that already exists, just add a number at the end 
of that username ,else print "Verified".'''
# Input Description:
'''First line consists of an integer N, denoting number of usernames. Second line consists of N spaced separated Strings, 
denoting usernames.'''
# Output Description:
# print the required output in a new line.

# Sample Input :
# 4
# abc aab abc aba
# Sample Output :
# Verified Verified abc1 aba

def unique_usernames(n, usernames):
    username_count = {}
    result = []
    
    for username in usernames:
        if username not in username_count:
            username_count[username] = 1
            result.append("Verified")
        else:
            new_username = f"{username}{username_count[username]}"
            username_count[username] += 1
            result.append(new_username)
    
    print(" ".join(result))

# Sample Input
n = int(input().strip())
usernames = input().strip().split()
unique_usernames(n, usernames)
```

```python
# 41
# Given a string two strings S1 and S2, remove characters from the S1 which are present in the S2.If S1 becomes empty then print -1
# Input Size : N <= 100000
# Sample Testcase :
# INPUT
# GUVI GEEK
# OUTPUT
# UVI

def remove_characters(s1, s2):
    # Create a set of characters to remove
    chars_to_remove = set(s2)
    # Build the resultant string without spaces between characters
    result = ''.join(char for char in s1 if char not in chars_to_remove)
    # Return the result or '-1' if the result is empty
    return result if result else '-1'

# Read input strings
s1, s2 = input().split()
# Print the result of the function
print(remove_characters(s1, s2))
```

```python
# 43
# Given a day, print 'yes' if it is a holiday otherwise print'no'.Assume that weekend days are holidays
# Sample Testcase :
# INPUT
# saturday
# OUTPUT
# yes
# INPUT
# monday
# OUTPUT
# no

def is_holiday(day):
    # Define a set of weekend days
    weekends = {'saturday', 'sunday'}
    
    # Check if the provided day is in the set of weekends
    if day.lower() in weekends:
        return 'yes'
    else:
        return 'no'

# Example usage
day = input().strip()
print(is_holiday(day))
```

```python
# 45
# Given a string S, print the encoded string by adding 3 to each character(a maps to d,b maps to e,c maps to f and so on).
# Input Size : 1 <= N <= 100000
# Sample Testcase :
# INPUT
# RADAR
# OUTPUT
# UDGDU

def caesar_cipher_shift_3(input_string):
    encoded_chars = []
    for char in input_string:
        if 'A' <= char <= 'Z':  # Uppercase letters
            shifted = chr((ord(char) - ord('A') + 3) % 26 + ord('A'))
        elif 'a' <= char <= 'z':  # Lowercase letters
            shifted = chr((ord(char) - ord('a') + 3) % 26 + ord('a'))
        else:
            shifted = char  # Non-alphabetic characters remain unchanged
        encoded_chars.append(shifted)
    return ''.join(encoded_chars)

# Example usage
input_string = input().strip()
print(caesar_cipher_shift_3(input_string))
```

```python
# 46
# Given a binary number convert it to hexadecimal.
# Sample Testcase :
# INPUT
# 1100100
# OUTPUT
# 64

def binary_to_hexadecimal(binary_string):
    # Convert binary string to an integer
    decimal_value = int(binary_string, 2)
    # Convert integer to hexadecimal string and remove the '0x' prefix
    hexadecimal_string = format(decimal_value, 'X')
    return hexadecimal_string

# Example usage
binary_string = input().strip()
print(binary_to_hexadecimal(binary_string))
```

```python
# 47
# Given a number N,check whether it has repeating digits in it.print 'yes' if it has repeating digits otherwise print 'no'.
# Sample Testcase :
# INPUT
# 11234
# OUTPUT
# yes

def has_repeating_digits(number):
    digits_seen = set()
    for digit in str(number):
        if digit in digits_seen:
            return 'yes'
        digits_seen.add(digit)
    return 'no'

# Example usage
number = int(input().strip())
print(has_repeating_digits(number))
```

```python
# 48
# Given a string S, print the reverse of the string after removing the vowels.If the resulting string is empty print '-1'.
# Input Size : 1 <= N <= 100000
# Sample Testcase :
# INPUT
# codekata
# OUTPUT
# tkdc

def remove_vowels_and_reverse(s):
    vowels = 'aeiouAEIOU'
    filtered_chars = [char for char in s if char not in vowels]
    if not filtered_chars:
        return '-1'
    return ''.join(filtered_chars)[::-1]

# Example usage
input_string = input().strip()
print(remove_vowels_and_reverse(input_string))
```

```python
# 50
# Print the position of first 1 from right to left, in binary representation of an Integer.
# Sample Testcase :
# INPUT
# 18
# OUTPUT
# 2

def position_of_rightmost_set_bit(n):
    if n == 0:
        return -1  # No set bits in zero
    position = 1
    while (n & 1) == 0:
        n >>= 1
        position += 1
    return position

# Example usage
number = int(input().strip())
print(position_of_rightmost_set_bit(number))
```
